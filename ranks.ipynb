```python
import pandas as pd
import tabula
import numpy as np
import os
import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
pd.set_option('display.max_rows', 15)
```

```python
file1 ="./HSS_1XX.pdf"
# print(os.path.isfile(file1))
tables = tabula.read_pdf(file1,pages='all',multiple_tables=False)
```

```python
dfs = []
for table in tables:
    df = pd.DataFrame(table)
    dfs.append(df)
```

```python
HSS = pd.concat(dfs,ignore_index=False)
HSS.columns=HSS.iloc[0]
HSS.drop(0,axis=0,inplace=True)
HSS.drop("Course Name",axis=1,inplace=True)
HSS.drop("Code",axis=1,inplace=True)
HSS
```

```python
names="hostels.pdf"
tables = tabula.read_pdf(names,pages='all',multiple_tables=False)
```

```python
namess=[]
for name in tables:
    df=pd.DataFrame(name)
    namess.append(df)
```

```python
df = pd.concat(namess, ignore_index=True)
indices = df[df['Remark'] == 'Remark'].index
df.drop(indices, inplace = True)
df.drop("Remark",axis=1,inplace=True)
df.drop("Hostel",axis=1,inplace=True)
df.drop("JEE Adv\rRoll No",axis=1,inplace=True)
```

```python
df['Branch'].unique()

```

```python
df['Branch'] = df['Branch'].map({
    'Biosciences and Bioengineering' : 'BSBE', 
    'Chemical Engineering' : 'CL', 
    'Chemical Science and Technology' : 'CST',
    'Civil Engineering' : 'CE',
    'Computer Science and Engineering' : 'CSE',
    'Data Science and Artificial Intelligence' : 'DSAI',
    'Electronics and Communication Engineering' : 'ECE',
    'Electronics and Electrical Engineering' : 'EEE',
    'Energy Engineering' : 'EE',
    'Engineering Physics' : 'EP',
    'Mathematics and Computing' : 'MNC',
    'Mechanical Engineering' : 'ME',
    'Design' : 'DOD'
    })
df.info()
df
```

```python
convert_dict = {
    'IITG Roll No' : int,
}
df = df.astype(convert_dict)
```

```python
# pd.set_option('display.max_columns', None)
# pd.set_option('display.max_rows', None)
# pd.set_option('display.max_colwidth', -1)
newNames=pd.read_csv("branchChange.txt",sep=' ')
newNames.drop("Name",axis=1,inplace=True)
newNames.drop("Present",axis=1,inplace=True)
newNames.drop("Sl.No",axis=1,inplace=True)
newNames.drop("New",axis=1,inplace=True)
newNames.rename(columns = {"NewRoll":"NewDept"}, inplace = True)
newNames.rename(columns = {"Unnamed: 6":"NewRoll"}, inplace = True)
newNames
```

```python
for index,newroll in newNames.iterrows():
    for index1,roll in df.iterrows():
        if(roll["IITG Roll No"]==newroll["Roll"]):
            df.loc[index1,"IITG Roll No"]=newroll["NewRoll"]
            df.loc[index1,"Branch"]=newroll["NewDept"]
            
```
